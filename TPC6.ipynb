**Resumo:** Este trabalho denominado *menu de manipulação de polinómios*, consiste num conjunto de funções que podem criar um polinómio; calcular: o seu grau, o polinómio num ponto e a sua derivada; simplificá-lo. Para além disso é possível apresentar a sua tabela e o seu gráfico. Este programa está escrito em **python**.

```python
import matplotlib.pyplot as plt

def menu():
    print(""""Bem vindo! 
          (1)Criar um polinómio                                    
          (2)Calcular grau do polinómio
          (3)Calcular polinómio num ponto
          (4)Calcular derivada do polinómio
          (5)Simplificar o polinómio 
          (6)Apresentar tabela do polinómio 
          (7)Ver gráfico do polinómio
          (0)Sair""")
    
    
    
def criarTermo (coef,exp):
    return (coef,exp)

def criarPolinomio ():   #criar polinómio vazio
    return []

def inserirTermo (p,t):
    p.append(t)  
    
def criarPolUtilizador():
    numtermos=int(input("Introduza o número de termos do polinómio."))
    p= criarPolinomio()
    for i in range(numtermos):
        coef=int(input("Introduza o coeficiente: "))
        exp=int(input("Introduza o expoente: "))
        t=criarTermo(coef,exp)
        inserirTermo(p,t)
    return p

def verTermo(t):
    c, e = t    
    if c==0 or e==0:
        return(str(c))        
    elif  c==1:
        return("x^" + str(e))
    elif e==1:
        return(str(c) + "x")
    else:
        return(str(c) + "x^" + str(e))
        
def verPolinomio(p):
    politexto=""
    if len(p)!=0:
        politexto = verTermo(p[0])
        for i in range(1, len(p)):
            if p[i][0] >= 0:
                politexto= politexto + " +" + verTermo(p[i])
            else:
                politexto= politexto + verTermo(p[i])
    return politexto
    
    
def grauPol(p):
    maior= p[0][1]
    for i in range(1, len(p)):
        if p[i][1]>maior:
                maior=p[i][1]
    return maior

def CalcPonto(p,x):
    total= 0
    for (c,e) in p:
        total= total + x*c **e 
    return total

def derivadaPol(p):
    derivada=[]
    for (c,e) in p:
        if e!= 0:
            derivada.append((c*e, e-1))
    return derivada

def tabela(p):
    linhas=int(input("Introduza o número de linhas que deseja: "))
    print("x" + " :: " + verPolinomio(p))
    for i in range(0,linhas):
        print(str(i) + " :: " + str(CalcPonto(p,i)))
        
def ordenarPol(p):
    p.sort(key=lambda tuplo: tuplo[1], reverse=True)
    return p

def simplicarPol(p):
    i=0
    while i <len(p)-1:
        if p[i][1]==p[i+1][1]:
            valor=p[i][0] + p[i+1][0]
            termo=criarTermo(valor,p[i][1])
            inserirTermo(p,termo)
            p.pop(i)
            p.pop(i)
            p=ordenarPol(p)    
        else:
            i+=1
    return p
            

def visualizarPol(p):
    x=[n for n in range(-100,100)]
    y=[]
    for n in x:
        y.append(CalcPonto(p,n))

    eixoXx =[x[0]-1,x[-1]+1]
    eixoXy =[0,0]

    plt.plot(eixoXx,eixoXy, color='red', linewidth = 1)

    eixoYx=[0,0]
    eixoYy=[min(y),max(y)]

    plt.plot(eixoYx,eixoYy, color='red', linewidth = 1)

    plt.title('Polinomio')
    
    plt.plot(x,y)

    plt.show()          

        
opcao=None
pol=None
while opcao!="0":
    menu()
    opcao=input("Introduza qual a opção\n")
    
    if opcao == "1":
        pol = criarPolUtilizador()
        print(verPolinomio(pol))
         
        
    elif opcao == "2":
        resultado = grauPol(pol)
        print("O grau do polinómio que inseriu é: ", resultado )
        
        
    elif opcao == "3":
        x=int(input("Indique o valor de x."))
        resultado = CalcPonto(pol,x)
        print( "O valor da ordenada correspondente à coordenada que inseriu é: ", resultado)
        
        
    elif opcao == "4":
        resultado = verPolinomio(derivadaPol(pol))
        print("A derivada do polinómio que inseriu é:", resultado)
        
        
    elif opcao == "5": 
        resultado = verPolinomio(simplicarPol(ordenarPol(pol)))
        print(resultado)
        
    elif opcao == "6":
        tabela(pol)
    
    elif opcao== "7":
        visualizarPol(pol)

    elif opcao=="0":
        print("Até à próxima!")
    else:
        print("Não introduziu nenhuma das opções!")
        
        

```
