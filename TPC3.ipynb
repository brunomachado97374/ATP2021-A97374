**Resumo:** Este trabalho denominado *menu*, consiste num conjunto de funções que podem gerar uma lista manualmente, uma lista aleatória, determinar o maior elemento da lista e organizar a lista de forma crescente. Este programa está escrito em **python**.

```python
def menu():
    print("""Bem vindo!
    (1) Criar uma lista manualmente;
    (2) Criar uma lista aleatória;
    (3) Ver o maior número da lista;
    (4) Bubble Sort;
    (0) Sair""")
    

def criarListaM():
    lista=[]
    tamanho=int(input("Introduza o número de elementos da lista.\n"))
    i=0
    while (len(lista)<tamanho):
        elemento=input("Introduza o valor do elemento " + str(i+1) +"\n")
        lista.append(elemento)
        i+=1
    return lista


from random import randrange
def criarListaAleatoria():
    lista=[]
    tamanho=int(input("Introduza o número de elementos da lista.\n"))
    i=0
    while (len(lista)<tamanho):
        elem=randrange(0,101)
        lista.append(elem)
        i+=1
    return lista

def numeroMaior(lista):
    maior=lista[0]
    for i in range(len(lista)):
        if lista[i]>maior:
            maior=lista[i]
    return maior


def bubbleSort(lista):
    paragem=False
    
    while paragem==False:
        
        paragem=True
        for i in range(len(lista)-1):

            if lista[i] > lista[i+1]:
                temp=lista[i]
                lista[i]=lista[i+1]
                lista[i+1]=temp
                paragem=False
    return lista


opcao=None
lst=None
while opcao!="0":
    menu()
    opcao=input("Introduza qual a opção\n")

    if opcao=="1":
        lst=criarListaM()
        print("A lista é: " + str(lst))

    elif opcao=="2":
            lst=criarListaAleatoria()
            print("A lista é: " + str(lst))

    elif opcao=="3":
        if lst:
            maior=numeroMaior(lst)
            print("O maior número é: " + maior) 
        else:
            print("Crie uma lista!")

    elif opcao=="4":

        if lst:
            listaOrdenada=bubbleSort(lst)
            print("A lista ordenada é: " + listaOrdenada)

        else:
            print("Crie uma lista!")

    elif opcao=="0":
        print("Obrigado")
        

    else:
        print("Opção não válida.")
       
   
    

    
    
    
    
    

```
