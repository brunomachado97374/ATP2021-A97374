**Resumo:** Este trabalho denominado *menu de manipulação de frações*, consiste num conjunto de funções que podem gerar uma fração manualmente ou de forma aleatória; simplicar frações; bem como operações básicas com funções, entre elas: soma, subtração, divisão e multiplicação. Este programa está escrito em **python**.

```python
import random

def menu ():
    print (""" Bem vindo!Selecione uma opção:
(1) Criar fração (manualmente ou aleatoriamente)
(2) Simplificar fração
(3) Somar
(4) Subtrair
(5) Multiplicar
(6) Dividir
(0) Sair 
           """)
           
def criarFracao ():
    op= input ("Pretende criar a função manualmente (M) ou aleatoriamente (A)? ")
    if op=="M":
        numerador= int (input("Insira o numerador da fração que pretende criar. "))
        denominador= int(input ("Insira o denominador da fração que pretende criar. "))
        if denominador== "0":
            print ("ERRO! O denominador da sua fração não pode ter o valor 0. ")
        return (numerador, denominador)
    elif op=="A":
        numerador= random.randint(1,100)
        denominador= random.randint(1,100)
    f=(numerador, denominador)
    return (f)
    
def verFracao(f):
    return (str(int(f[0]))+"/"+str(int((f[1]))))

def mdc(a,b):
    if a%b==0:
        return b
    elif b%a==0:
        return a
    else:
        if a>b:
            return mdc(b, a%b)
        else:
            return mdc(a, b%a)
    
def simplificarFracao (f):
    num, denom = f
    c= mdc (num, denom)
    return (int(num/c), int(denom/c))

def somarFrac (f1,f2):
    denom= f1[1] * f2 [1]
    num= f1[0] * f2[1] + f2[0] * f1[1]
    j= (num, denom)
    return simplificarFracao(j)

def subtrairFrac (f1,f2):
    denom= f1[1] * f2 [1]
    num= f1[0] * f2[1] - f2[0] * f1[1]
    j= (num, denom)
    return simplificarFracao(j)

def multiplicarFrac (f1,f2):
    denom= f1[1] * f2 [1]
    num= f1[0] * f2[0] 
    j= (num, denom)
    return simplificarFracao(j)


def dividirFrac (f1,f2):
    denom= f1[0] * f2 [1]
    num= f1[1] * f2[0] 
    j= (num, denom)
    return simplificarFracao(j)

op="6"
menu()
while op!="0":
    op= input ("Qual é a opção que pretende selecionar? ")
    if op=="1":
        fraca= criarFracao()
        print (verFracao(fraca))
        
    elif op=="2":
        try:
            fracao= verFracao ( simplificarFracao(fraca))
            print(fracao)
        except:
            print ("Ainda não criou uma fração. ")
        
    elif op=="3":
        try:
            frac1= fraca
            print ("Já tem uma fração criada. Vai agora criar a segunda fração que quer somar.")
            frac2= criarFracao()
            print (verFracao(frac2))
            fraca = somarFrac(frac1, frac2)
            print (verFracao(fraca))
        except:
            print ("Ainda não criou uma fração para somar com a que irá criar. ")
        
    elif op=="4":
        try:
            frac1= fraca
            print ("Já tem uma fração criada. Vai agora criar a segunda fração que quer subtrair.")
            frac2= criarFracao()
            print (verFracao(frac2))
            fraca = subtrairFrac(frac1, frac2)
            sub= verFracao(fraca)
            print (sub)
        except:
            print("Ainda não criou uma fração para subtrair com outra que irá criar. ")
            

    elif op=="5":
        try:
            frac1= fraca
            print ("Já tem uma fração criada. Vai agora criar a segunda fração que quer multiplicar.")
            frac2= criarFracao()
            print (verFracao(frac2))
            fraca = multiplicarFrac(frac1, frac2)
            mul= verFracao(fraca)
            print (mul)
        except:
            print("Ainda não criou uma fração para multiplicar com outra que irá criar. ")
        
    elif op=="6":
        try:
            frac1= fraca
            print ("Já tem uma fração criada. Vai agora criar a segunda fração que quer dividir.")
            frac2= criarFracao()
            print (verFracao(frac2))
            fraca = dividirFrac(frac1, frac2)
            div= verFracao(fraca)
            print (div)
        except:
            print("Ainda não criou uma fração para dividir com outra que irá criar. ")
        
    elif op=="0":
        print ("Até à próxima! ")
    else:
        print ("Opção inválida! ")
```
